```python
def age_group (age_cert):
    if age_cert in['TV-MA', 'R','NC-17']:
        return "Adult"
    elif age_cert in['PG', 'TV-14', 'G', 'PG-13', 'TV-PG']:
        return "Requires supervision"
    elif age_cert in ['TV-Y','TV-G', 'TV-Y7']:
        return "Good for children"
    elif age_cert in ['TV-Y', 'TV-G', 'TV-Y7']:
        return "Good for children"
    else:
        return"Not Rated"
    
```

```python
df["age_group_rating"] = df ["age_certification"]. apply (age_group)
```

4.show barchart for movie genres for children movies

```python

q4_df = df.copy()

q4_df['genres'] = q4_df['genres'].str.split(",")
children_movies = q4_df[df["age_group_rating"]=="Good for children"]
children_movies
#count the occurence of each genre
genre_counts = children_movies['genres'].apply(pd.Series).stack().value_counts()

#plot the bar chart
genre_counts.plot(kind='bar')
plt.title('Genres for Children')
plt.xlabel('Genre')
plt.ylabel('Number of Titles')
plt.show()
```
